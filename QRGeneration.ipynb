# QR Code for London and Parish with Python

## (Input: Any 2 famous Location like Taj Mahal, Eiffel Tower, etc..., Output: QR Code)

```python
# Required Imports
import cv2
import qrcode
import matplotlib.pyplot as plt
```

```python
#Showing the London Image
image = cv2.imread('london.jpg')
plt.imshow(image)
plt.title("London Image")
plt.show()
```

```python
# Making the QR
london = qrcode.make(image)
```

```python
# Saving the QR Image
london.save('londonQR.jpg')
```

```python
#Showing the London QR Generated
qr = cv2.imread('londonQR.jpg')
plt.imshow(qr)
plt.title("London QR")
plt.show()
```

```python
#Showing the Paris Image
image = cv2.imread('paris.jpg')
plt.imshow(image)
plt.title("Paris Image")
plt.show()
```

```python
# making the QR
paris = qrcode.make(image)
```

```python
# Saving the QR Image
paris.save('parisQR.jpg')
```

```python
#Showing the Paris QR Generated
qr = cv2.imread('parisQR.jpg')
plt.imshow(qr)
plt.title("Paris QR")
plt.show()
```
